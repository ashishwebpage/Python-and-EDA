```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

- Here %matplotlib inline helps while plotting. We don't need to write plt.show() all the time.

### Scatter plot

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
# Scatter plot using matplotlib
# Scatter plot plots a straight line i.e it plots the point only and doesn't show the line connecting the points
```

```python
plt.scatter(x,y,c='r')
# Here x and y must be of same size
plt.xlabel('X axix')
plt.ylabel('Y axix')
plt.title('Simpel 2D Graph')
#plt.savefig('Name') is used to save the graph 
plt.savefig('fig.jpg')
#plt.savefig('fig.jpeg')
#plt.savefig('fig.png')
```

### Plt Plot

```python
y=x*x
```

```python
#plt.plot(x,y,'r*--')
# plt.plot(x,y,'ro--')
# plt.plot(x,y,'ro')
# plt.plot(x,y,'r')
# plt.plot(x,y,'r--')
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=10)
plt.xlabel('X axix')
plt.ylabel('y axix')
plt.title('Basic line graph')
```

- Here marker size gives the size of the points

### Creating subplots

- subplot can be created using plt.subplot(no of rows, no of columns, position which starts from 1)

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
plt.subplot(2,2,3)
plt.plot(x,y,'bo--')
plt.subplot(2,2,4)
plt.plot(x,y,'y*--')
```

### Plotting sine curve

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine waveform')
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sine wave')


plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('Cosine wave')
```

### Bar plot

```python
x=[5,10,15]
y=[10,15,20]

x1=[20,25,30]
y1=[25,30,35]
plt.bar(x,y,)
plt.bar(x1,y1,color='g')
plt.title("Bar Graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

### Histograms 
- Histograms represent the density or the number of occurance of numbers over a specified range such as 0-10, 10-20

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,bins=30)
# Here bins indicates the number in which the x axis is divided
# Increasing the number of bins means shortening the range of numbers
plt.title("Histogram")
```

### Box plot using matplotlib

- Here the circles outside the box plot are the outliers
- Due to the range (1,5), we obtain 4 lists of values of data type list
- vert=true makes the boxplot vertical and patch_artist=true makes the box plot colorful

```python
data=[np.random.normal(0,std,100) for std in range(1,5)]
plt.boxplot(data,vert=True,patch_artist=True);
```

### Pie Chart

```python
labels='c++','python','java','ruby'
sizes=[215,130,240,210]
colors=['gold','yellow','blue','purple']
explode=(0.2,0.1,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.3f%%',shadow=True)
# Here 1.3f represents the presence of the 3 decimal palces
plt.axis('equal')
```
